```python
df['category_id'] = df['offense_tag'].factorize()[0]
category_id_df = df[['offense_tag', 'category_id']].drop_duplicates()
```

```python
category_to_id = dict(category_id_df.values)
id_to_category = dict(category_id_df[['category_id', 'offense_tag']].values)
df.head()
```

```python
# Data exploration and visualization
fig = plt.figure(figsize=(8,6))
colors = ['grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey','grey']
df.groupby('offense_tag').grievance.count().sort_values().plot.barh(
    ylim=0, color=colors, title= 'Number of Grievance in Each Offenses')
plt.xlabel('Number of ocurrences', fontsize = 20);
```

```python
# Text preprocessing and feature extraction
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, ngram_range=(1, 2), stop_words='english')
features = tfidf.fit_transform(df.grievance).toarray()
labels = df.category_id
print("Each of the %d complaints is represented by %d features (TF-IDF score of unigrams and bigrams)" %(features.shape))
```

```python
N = 5
for offense_tag, category_id in sorted(category_to_id.items()):
  features_chi2 = chi2(features, labels == category_id)
  indices = np.argsort(features_chi2[0])
  feature_names = np.array(tfidf.get_feature_names_out())[indices]
  unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
  bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
  print("n==> %s:" %(offense_tag))
  print("  * Most Correlated Unigrams are: %s" %(', '.join(unigrams[-N:])))
  print("  * Most Correlated Bigrams are: %s" %(', '.join(bigrams[-N:])))
```

```python
# Train and evaluate the model
X = df['grievance']
y = df['offense_tag']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=1)

models = [
    LinearSVC()
]
# 5 Cross-validation
CV = 5
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []
for model in models:
  model_name = model.__class__.__name__
  accuracies = cross_val_score(model, features, labels, scoring='accuracy', cv=CV)
  for fold_idx, accuracy in enumerate(accuracies):
    entries.append((model_name, fold_idx, accuracy))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])

# Initialize and train the LinearSVC model
model = LinearSVC()
model.fit(tfidf.transform(X_train), y_train)

# Predict on the test set
y_pred = model.predict(tfidf.transform(X_test))

# Print classification report and confusion matrix
print('Offense Classification Metrics')
print(metrics.classification_report(y_test, y_pred, target_names=df['offense_tag'].unique()))

conf_mat = confusion_matrix(y_test, y_pred)
# ... (your confusion matrix visualization code here)

fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(conf_mat, annot=True, cmap="Blues", fmt='d',
            xticklabels=category_id_df.offense_tag.values,
            yticklabels=category_id_df.offense_tag.values)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title("CONFUSION MATRIX - Linear SVC", size=20);
```

```python
# Predict offenses for a given complaint text
complaint = "I want to report an incident regarding of stealing incident that happened on the school canteen office where some of my valuable items like my money and laptop has been steal by someone."
predicted_offenses = model.predict(tfidf.transform([complaint]))

# Get the top 10 predicted offenses' category IDs
top_10_offense_ids = model.classes_[np.argsort(model.decision_function(tfidf.transform([complaint])))[0, -15:]]

print("Predicted Offense/s:", predicted_offenses)
print("Top 10 Possible Offenses:", top_10_offense_ids)
```
